```python
from bs4 import BeautifulSoup
import requests
```

```python
url = ('https://en.m.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue')
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[2]
```

```python
print(table)
```

```python
titles = table.find_all('th')
```

```python
print(titles)
```

```python
world_table_titles = [title.text.strip() for title in titles]
```

```python
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df  = pd.DataFrame(columns = world_table_titles)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
print(column_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    print(individual_data)

    length = len(df)
    df.loc[length] =  individual_data
```
